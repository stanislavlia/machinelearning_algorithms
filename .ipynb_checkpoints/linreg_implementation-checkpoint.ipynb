```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class myLinearRegression():
    def __init__(self):
        weights = None
        bias = None
        
        self.X = None
        self.Y = None
        self.MSE_history = []
        
    def _MSE(self, Y_est):
        
        MSE = (1 / len(Y_est)) * np.sum((Y_est - self.Y) ** 2)
        return MSE
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
    
    def compute_gradients(self, Y_est):
        
        errors = (Y_est - self.Y)
        n = self.X.shape[0]
        
        W_gradients = (2 / n) * np.dot(self.X.T, errors)
        b_gradient  = (2 / n) * np.sum(errors)
        
        return W_gradients, b_gradient
    
    def fit(self, X, Y, alpha=0.001, epochs=100):
        
        self.X = X
        self.Y = Y
        print( "X_shape ",self.X.shape)
        print( "Y_shape ",self.Y.shape)
        
        
        #intialize weights
        self.weights = np.zeros((X.shape[1],1))
        self.MSE_history = []
        
        print("Weights shape :", self.weights.shape)
        self.bias = np.random.random()
        
        for epoch_num in range(1, epochs+1):
            
            Y_est = self.predict(self.X)
            
            MSE = self._MSE(Y_est)
            self.MSE_history.append(MSE)
            
            #setting step to display MSE
            epoch_step = epochs // 10
            
            
            if (epoch_num % epoch_step) == 0:
                
                print("Epoch num {}; MSE = {}".format(epoch_num, MSE))
            
            w_grads, b_grad = self.compute_gradients(Y_est)

            #correct weights
            self.weights = self.weights - alpha * w_grads
            self.bias = self.bias - alpha * b_grad
            
        print(self.weights, self.bias)
        return (self.weights, self.bias, self.MSE_history)
            
        
        
    
    
```

```python
#Trying to use the model

import pandas as pd
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("multiple-linear-regression-dataset.csv")

data.head()
```

```python
deneyim = data["deneyim"].values.reshape(-1, 1)
maas = data["maas"].values.reshape(-1, 1)

X = np.column_stack([deneyim, maas])
y = data["yas"].values.reshape(-1, 1)

stdscal = StandardScaler()
X_scaled = stdscal.fit_transform(X)

```

```python
linreg = myLinearRegression()
_, _, mse_history= linreg.fit(X_scaled, y, 0.1, 200)
_, _, mse_history2= linreg.fit(X_scaled, y, 0.01, 200)
```

```python

y_est = linreg.predict(X_scaled)

y_compare = pd.DataFrame({"Y_est" : y_est.flatten(), "Y_real" : y.flatten()})
y_compare
```

```python
#let's plot our algorithm was learning

plt.figure(figsize=(8, 6))

iterations = np.arange(0, len(mse_history), 1)

plt.plot(iterations, mse_history, label = "alpha = 0.1")
plt.plot(iterations, mse_history2, label = "alpha = 0.01")

plt.title("Minimizing Cost function")
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("MSE")
```
